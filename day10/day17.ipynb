```python
import matplotlib.pyplot as plt
import numpy as np
```

# Types of Graphs
1. plot -            line graph, between two parameters
2. bar -             bar graph of single parameter( optional cal_data )
3. hist -            histogram of a bunch of data i.e., frequency of single parameter
4. scatter -         dot graph between parameters
5. boxplot / box -   statistical overview in form of box for a single parameter
                     mean, outliers, quatities (25%, 75%), IOR
                     (optional-categorical data)
5. stem -            more like bar graph, lines instead of bars
6. violinpilot -     violin shaped statistical overview graph 
                     spread of data (more like hist) 
                     single parameter (optional cal_data)

```python
plt.bar(x,y)
plt.title('Bar Graph')
plt.show()

plt.hist(x)
plt.title('Hist')
plt.show()

plt.scatter(x,y)
plt.title('Scatter')
plt.grid()
plt.xlabel('x -- >')
plt.ylabel('y -- >')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
y2=x**4
```

```python

plt.plot(x, y, x, y2)
plt.legend(['y=x**3','y=x**4'])
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(x,y, lw=3, marker='o',markerfacecolor='hotpink',markersize=10)
plt.plot(x,y2, marker='v', markersize=10)
plt.legend(['y=x**3','y=x**4'])
plt.grid()
plt.show()
```
